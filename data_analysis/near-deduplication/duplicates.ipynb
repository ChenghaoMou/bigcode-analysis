```python
from datasets import load_dataset, load_from_disk
import os
import pandas as pd
from IPython.display import display, Markdown
```

```python
%%capture --no-display

ds = load_dataset(
    "codeparrot/codeparrot-clean-valid",
    split="train",
    num_proc=os.cpu_count(),
)

ds = ds.map(
    lambda _, idx: {"__id__": idx},
    with_indices=True,
    num_proc=os.cpu_count(),
    batched=True,
    desc="Adding unique index...",
)
```

```python
neighbors = load_from_disk("results/codeparrot/codeparrot-clean-valid/default/80/all/train/content/neighbors")
```

```python
records = neighbors.filter(lambda x: x["__neighbors__"]).select(range(10))
display(Markdown(pd.DataFrame(records)[["__id__", "__neighbors__"]].to_markdown()))
```
